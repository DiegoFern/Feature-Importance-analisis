# Fake importances:

It's usual use the features importance to evaluate and trust in the differents variables, but it's not always trusted.

In this experiment we will train a GradientBoostingClassifier with a set. After that we will put another random variable and we will see the importances:

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_iris,load_diabetes
import pandas as pd
from sklearn import tree
from sklearn import ensemble
```

```python
data = load_diabetes()
```

```python
P = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
target = pd.Series(data.target)
```

```python
clf = ensemble.GradientBoostingClassifier().fit(P,target)
```

```python
clf.feature_importances_
```

```python
P.head()
```

```python
pd.Series(clf.feature_importances_,index=list(P)).sort_values().plot.barh()
```

After that we will append a fake variable with a normal distribution:

```python
P['fake']=pd.np.random.random(size=P.shape[0])
```

```python
P
```

```python
clf_fake =ensemble.GradientBoostingClassifier(random_state=1242).fit(P,target)
clf_fake.feature_importances_
```

Incredible, the most important is variable is the fake one!

```python

pd.Series(clf_fake.feature_importances_,index=list(P)).sort_values().plot.barh()
```
